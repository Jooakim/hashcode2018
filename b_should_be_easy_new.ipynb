```python
def locate_min(a):
    smallest = min(a)
    return smallest, [int(index) for index, element in enumerate(a) if smallest == element]
```

```python
df = pd.read_csv('b_should_be_easy.in', delimiter=' ')
data = df.columns
df.columns = ['a', 'b', 'x', 'y', 's', 'f']
df




start_pos = np.asanyarray([df['a'].values, df['b'].values]).T
end_pos = np.asanyarray([df['x'].values, df['y'].values]).T
tmp_distances = start_pos - end_pos
scalar_distances = np.sum(abs(tmp_distances), axis=1)

earliest_start = df['a'].values
latest_finish = df['f'].values

allowed_lateness = latest_finish - earliest_start - scalar_distances


n_cars = int(data[3])
T = int(data[5])

car_trips = []

for c in range(n_cars):
    car_pos = [0, 0]
    t = 0
    
    current_car_trip = []
    print("Car {0}".format(c))
    while t < T:
        if np.max(start_pos[:,0]) < 0:
            break
            
        tmp_start_dist = car_pos - start_pos
        scalar_start_dist = np.sum(np.abs(tmp_start_dist), axis=1)
        loss = t + scalar_start_dist - earliest_start
        
        valid_trips = [i for i,x in enumerate(loss) if x <= allowed_lateness[i]]
        if len(valid_trips) == 0:
            break
        valid_trips_trip_ind = locate_min(np.abs(loss[valid_trips]))[1]
        
        candidate_trips_ind = [valid_trips[x] for x in valid_trips_trip_ind]
        #candidate_trips_ind = valid_trips[valid_trips_trip_ind]

        trip_ind = np.argmax(scalar_distances[candidate_trips_ind])
        
        chosen_trip_ind = candidate_trips_ind[trip_ind]
        
        current_car_trip.append(chosen_trip_ind)
        
        start_pos[chosen_trip_ind, 0] = -10000000000
        
        # Update time
#        print(scalar_distances[chosen_trip_ind], scalar_start_dist[chosen_trip_ind])
 #       print()
        t = t + scalar_distances[chosen_trip_ind] + scalar_start_dist[chosen_trip_ind]
        if loss[chosen_trip_ind] < 0:
            t = t + np.abs(loss[chosen_trip_ind])
        #print(t)
        # Update car pos
        car_pos = end_pos[chosen_trip_ind]
        
    # Append to car trips
    car_trips.append(current_car_trip)

car_trips


    
```
